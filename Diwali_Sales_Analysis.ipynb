### Gender

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# Use font that supports ₹ symbol
plt.rcParams['font.family'] = 'DejaVu Sans'

# Grouped data
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size
plt.figure(figsize=(8, 6))

# Custom colors for M and F
palette = {'M': '#1f77b4', 'F': '#ff69b4'}

# Barplot with hue to avoid deprecation
ax = sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender', palette=palette, legend=False)

# Add ₹ labels on top of bars
for bar in ax.containers:
    ax.bar_label(bar, fmt='₹%.0f', fontsize=12, padding=3)

# Titles and formatting
plt.title('Total Sales Amount by Gender', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Total Amount Spent (₹)', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
import matplotlib.pyplot as plt
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# Total Amount vs Age Group
import matplotlib.pyplot as plt
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
plt.show()
```

### State

```python
# total number of orders from top 10 states
import matplotlib.pyplot as plt
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
plt.show()
```

```python
# total amount/sales from top 10 states
import matplotlib.pyplot as plt
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
# Set plot size before plotting
sns.set(rc={'figure.figsize': (7, 5)})

# Fix: Add hue and disable legend to avoid deprecation warning
ax = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status', palette='Set2', legend=False)

# Add count labels
for bars in ax.containers:
    ax.bar_label(bars, fontsize=12)

# Optional formatting
plt.title('Distribution by Marital Status', fontsize=16)
plt.xlabel('Marital Status', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Unmarried', 'Married'], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.show()
```

### Product Category

```python
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show() 
```

```python
import matplotlib.pyplot as plt
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
plt.show()
```

```python
import matplotlib.pyplot as plt
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
plt.show()
```

```python
# Prepare data: Top 10 most sold products by total Orders
top10_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 7))
sns.set_style('whitegrid')

ax = top10_products.plot(kind='bar', color='skyblue', edgecolor='black')

# Add value labels
for i, val in enumerate(top10_products):
    ax.text(i, val + 1, str(int(val)), ha='center', va='bottom', fontsize=10)

# Formatting
plt.title('Top 10 Most Sold Products by Order Volume', fontsize=18, fontweight='bold')
plt.xlabel('Product ID', fontsize=14)
plt.ylabel('Total Orders', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
